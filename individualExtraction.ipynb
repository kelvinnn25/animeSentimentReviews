```python
import time

def get_reviews(df, pages_per_link=5, delay_seconds=1):
    # Create a new DataFrame to store reviews
    reviews_df = pd.DataFrame(columns=['title', 'link', 'username', 'tags', 'review'])

    for index, row in df.iterrows():
        for page in range(1, pages_per_link + 1):
            # Modify the link by adding 'reviews?spoiler=on&p={page}' at the end
            modified_link = f"{row['link']}/reviews?spoiler=on&p={page}"
            print(f"Processing link: {modified_link}")

            # Make a request to the modified link
            response = requests.get(modified_link)

            # Check if the request was successful (status code 200)
            if response.status_code == 200:
                print(f"Successfully fetched page {page}")
                # Parse the HTML content of the page
                soup = BeautifulSoup(response.content, 'html.parser')

                # Select review elements on the page
                review_elements = soup.select('.review-element.js-review-element')

                # Extract and store each review separately
                for element in review_elements:
                    # Extract username, tags, and review text
                    tags_element = element.select_one('.tags')
                    review_text_element = element.select_one('.text')


                    tags = tags_element.text.strip()
                    review_text = review_text_element.text.strip() if review_text_element else None

                    reviews_df = reviews_df.append({
                        'title': row['title'],
                        'link': row['link'],
                        'tags': tags,
                        'review': review_text
                    }, ignore_index=True)
                # Introduce a delay before the next request
                time.sleep(delay_seconds)

            else:
                print(f"Failed to fetch page {page}. Status code: {response.status_code}")

    return reviews_df

# Assuming your original DataFrame is named 'top50_df'
# Call the function to get and add reviews from 25 pages per link with a delay of 1 second
reviews_df = get_reviews(animeDf1, pages_per_link=5, delay_seconds=1)
```

```python
def get_failed_reviews(links, delay_seconds=1):
    # Create a new DataFrame to store reviews
    reviews_df = pd.DataFrame(columns=['title', 'link', 'tags', 'review'])

    for link in links:

          current_link = f"{link}"
          print(f"Processing link: {current_link}")

          # Make a request to the current link
          response = requests.get(current_link)

          # Check if the request was successful (status code 200)
          if response.status_code == 200:
              print(f"Successfully fetched link {link}")
              # Parse the HTML content of the page
              soup = BeautifulSoup(response.content, 'html.parser')

              # Select review elements on the page
              review_elements = soup.select('.review-element.js-review-element')

              # Extract and store each review separately
              for element in review_elements:
                  # Extract username, tags, and review text
                  tags_element = element.select_one('.tags')
                  review_text_element = element.select_one('.text')

                  tags = tags_element.text.strip()
                  review_text = review_text_element.text.strip() if review_text_element else None

                  reviews_df = reviews_df.append({
                      'title': link.split('/')[-2],  # Extract title from the link
                      'link': link,
                      'tags': tags,
                      'review': review_text
                  }, ignore_index=True)
              # Introduce a delay before the next request
              time.sleep(delay_seconds)

          else:
              print(f"Failed to fetch link {link}. Status code: {response.status_code}")

    return reviews_df
```
